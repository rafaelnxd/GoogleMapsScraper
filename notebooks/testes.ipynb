```python
import pandas as pd
import folium
from folium.plugins import HeatMap

data = pd.read_csv('combined_maps_data.csv')  

min_latitude = -23.791
max_latitude = -23.453
min_longitude = -46.819
max_longitude = -46.338

filtered_data = data[(data['latitude'] >= min_latitude) &
                     (data['latitude'] <= max_latitude) &
                     (data['longitude'] >= min_longitude) &
                     (data['longitude'] <= max_longitude)]

average_lat = (min_latitude + max_latitude) / 2
average_lon = (min_longitude + max_longitude) / 2
m = folium.Map(location=[average_lat, average_lon], zoom_start=14)
locations = filtered_data[['latitude', 'longitude']].dropna()  
heat_data = locations.values.tolist()
HeatMap(heat_data).add_to(m)
m.save('heatmap.html')
```
